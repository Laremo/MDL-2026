# PRACTICA No. 2. NEURONA DE McCulloch Pitts.
 
El modelo de McCulloch-Pitts (MCP) es una simplificación de una neurona, que toma varias entradas, las suma con un peso, y luego aplica una función de activación (generalmente una función escalón) para determinar la salida.

En esta practica entrenaremos una célula de McCulloch-Pitts para que sea capaz de comportarse como una compuerta lógica AND de dos entradas.

### Paso 1: definir librerias necesarias

```python
import numpy as np
```

### Paso 2; Definir la fucnión de activación para el AND lógico

```python
def mcculloch_pitts_AND(x1, x2):
    # Pesos de las entradas (se mantienen iguales)
    w1, w2 = 1, 1 

    # Umbral: AUMENTA para ser más restrictivo
    teta = 2

    # Suma ponderada de las entradas
    weighted_sum = x1 * w1 + x2 * w2 

    # Salida de la célula
    if weighted_sum >= teta:
        return 1
    else:
        return 0
```

### Paso 3: Probar la neurona de McCullotch-Pitts con el AND lógico

```python
# Ejemplo de uso
print(mcculloch_pitts_AND(0, 0))  
print(mcculloch_pitts_AND(0, 1)) 
print(mcculloch_pitts_AND(1, 0)) 
print(mcculloch_pitts_AND(1, 1)) 
```
