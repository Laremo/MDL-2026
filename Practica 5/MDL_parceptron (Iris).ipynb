## Practica 6: Perceptron de Frank Rosenbaltt para clasificar la flor de Iris

Un perceptrón es un tipo de modelo de red neuronal muy simple, utilizado principalmente para problemas de clasificación binaria. Fue desarrollado en 1958 por Frank Rosenblatt y es considerado uno de los primeros algoritmos de aprendizaje supervisado.

El modelo toma un conjunto de entradas, las procesa mediante una serie de operaciones matemáticas y produce una salida que puede ser 0 o 1, dependiendo de si el resultado final supera un umbral predeterminado.
